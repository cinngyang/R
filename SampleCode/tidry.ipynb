+ key 的命名轉成長數據欄位 , value 欄位命名

```r
stocks <- data.frame(
  time = as.Date('2009-01-01') + 0:9,
  X = rnorm(10, 0, 1),
  Y = rnorm(10, 0, 2),
  Z = rnorm(10, 0, 4)
)
stocks
```

```r
stocksm <- stocks %>% gather(stock, price, -time)
stocksm[1:3,]
```

```r
stocksm %>% spread(stock, price)
```
