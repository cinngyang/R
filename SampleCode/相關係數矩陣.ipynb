```r
library(dplyr)
library(corrplot)
library(ggplot2)
```

```r
MONO=read.csv("C:/Data/GitHub/Python/SampleFile/MONO.csv")
MixWT=read.csv("C:/Data/GitHub/Python/SampleFile/MixWT.csv")
```

```r
P3806<-filter(MixWT,PRODUCTNO=='95.S3806.W0A0C8Z')
P3806=left_join(P3806,MONO,by=c('MONO'='MONO'))
```

```r
P3806Cor=filter(P3806,WDD=='452.5-450')%>%
select('WT','SPC_VALUE','SPC_RANGE','PRE_CIE_X_AVG','PRE_CIE_Y_AVG','CIE_X_AVG','CIE_Y_AVG')

P3806Cor$PRE_CIE_X_AVG<-as.numeric(P3806Cor$PRE_CIE_X_AVG)
P3806Cor$PRE_CIE_Y_AVG<-as.numeric(P3806Cor$PRE_CIE_Y_AVG)
P3806Cor$CIE_X_AVG<-as.numeric(P3806Cor$CIE_X_AVG)
P3806Cor$CIE_Y_AVG<-as.numeric(P3806Cor$CIE_Y_AVG)
P3806Cor$WT<-as.numeric(P3806Cor$WT)

P3806Cor<-cor(P3806Cor)
corrplot(P3806Cor, method="number")
```
